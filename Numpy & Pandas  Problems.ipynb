Q1. Create a CSV file containing product names and prices. Read the data using
Pandas. Use NumPy to calculate the average price. Write a function to identify
products priced above the average and display them.

Concepts Used:

 Pandas DataFrame

 NumPy

 Functions

 Control Statements

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("/content/product price .csv")
```

```python
df
```

```python
print(df['Price'].mean())
```

```python
avg_price = df['Price'].mean()
```

```python
def above_average_products(df, avg_price):
    print("\nProducts above average price:")

    for index, row in df.iterrows():
        if row["Price"] > avg_price:
            print(row["Product"], "-", row["Price"])
```

```python
above_average_products(df, avg_price)
```

 Q2. Read daily temperature data from a CSV file using Pandas. Use NumPy to
calculate the highest, lowest, and average temperatures. Write functions to classify
each day as:

 Cold (&lt;20°C)

 Moderate (20–30°C)

 Hot (&gt;30°C)

Display the category for each day.
Concepts Used:

 Pandas

 NumPy

 Functions

 if-elif-else

```python
import pandas as pd

data = {
    "Day": ["Monday", "Tuesday", "Wednesday", "Thursday", "Friday", "Saturday", "Sunday"],
    "Temperature": [18, 25, 32, 28, 15, 35, 22]
}

df = pd.DataFrame(data)

print(df)
```

```python
import numpy as np

highest = np.max(df["Temperature"])
lowest = np.min(df["Temperature"])
average = np.mean(df["Temperature"])

print("Highest Temperature:", highest)
print("Lowest Temperature:", lowest)
print("Average Temperature:", average)
```

```python
def classify_temperature(temp):
    if temp < 20:
        return "Cold"
    elif temp <= 30:
        return "Moderate"
    else:
        return "Hot"
```

```python
for index, row in df.iterrows():
    print(row["Day"], ":", classify_temperature(row["Temperature"]))
```
