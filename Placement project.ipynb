# # importing label encoder

```python
from sklearn.preprocessing import LabelEncoder
```

```python
le = LabelEncoder()
```

```python
for col in df.select_dtypes(include="object").columns:
    df[col] = le.fit_transform(df[col])
```

```python
df.head()
```

```python
X = df.drop(["StudentID", "PlacementStatus"], axis=1)
y = df["PlacementStatus"]
```

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
X_scaled = pd.DataFrame(
    scaler.fit_transform(X),
    columns=X.columns
)

print(X_scaled.head())
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y, test_size=0.2, random_state=42
)
```

```python
X_scaled
```

```python
X_train
```

```python
X_test
```

```python
from sklearn.linear_model import LogisticRegression

model = LogisticRegression()
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
y_pred
```

# # evaluating the results

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

print("Accuracy:", accuracy_score(y_test, y_pred))
print(confusion_matrix(y_test, y_pred))
print(classification_report(y_test, y_pred))
```

# KNN

```python
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score

for k in range(1, 21):
    knn = KNeighborsClassifier(n_neighbors=k)
    knn.fit(X_train, y_train)

    y_pred = knn.predict(X_test)

    acc = accuracy_score(y_test, y_pred)

    print("K =", k, "Accuracy =", acc)
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.neighbors import KNeighborsClassifier

params = {
    'n_neighbors': range(1, 31)
}

grid = GridSearchCV(
    KNeighborsClassifier(),
    param_grid=params,
    cv=5
)

grid.fit(X_train, y_train)
```

```python
print("Best K:", grid.best_params_)
print("Best Score:", grid.best_score_)
```

```python
# Predict using the best K (29) found by GridSearch
y_pred_knn = grid.predict(X_test)
```

```python
y_pred_knn
```

```python

from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

print("--- KNN Test Performance (K=29) ---")
print("Accuracy:", accuracy_score(y_test, y_pred_knn))
print("\nConfusion Matrix:\n", confusion_matrix(y_test, y_pred_knn))
print("\nClassification Report:\n", classification_report(y_test, y_pred_knn))
```

# Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier

dt = DecisionTreeClassifier(random_state=42)

dt.fit(X_train, y_train)
```

```python
y_pred_dt = dt.predict(X_test)
```

```python
y_pred_dt
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# Accuracy
accuracy = accuracy_score(y_test, y_pred_dt)
print("Accuracy:", accuracy)

# Confusion Matrix
cm = confusion_matrix(y_test, y_pred_dt)
print("\nConfusion Matrix:")
print(cm)

# Classification Report
report = classification_report(y_test, y_pred_dt)
print("\nClassification Report:")
print(report)
```

# # Random forest

```python
from sklearn.ensemble import RandomForestClassifier

rf = RandomForestClassifier(
    n_estimators=200,
    max_depth=10,
    random_state=42
)


rf.fit(X_train, y_train)
```

```python
y_pred_rf = rf.predict(X_test)
```

```python
y_pred_rf
```

```python
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

# Accuracy
accuracy = accuracy_score(y_test, y_pred_rf)
print("Accuracy:", accuracy)

# Confusion Matrix
cm = confusion_matrix(y_test, y_pred_rf)
print("\nConfusion Matrix:")
print(cm)

# Classification Report
report = classification_report(y_test, y_pred_rf)
print("\nClassification Report:")
print(report)
```

```python
lr_acc = 0.7945
knn_acc = 0.791
dt_acc = 0.725
rf_acc = 0.7885

print("Logistic Regression:", lr_acc)
print("KNN:", knn_acc)
print("Decision Tree:", dt_acc)
print("Random Forest:", rf_acc)
```

```python
import matplotlib.pyplot as plt

results = pd.DataFrame({
    "Algorithm": [
        "Logistic Regression",
        "KNN",
        "Decision Tree",
        "Random Forest",
    ],
    "Accuracy": [lr_acc, knn_acc, dt_acc, rf_acc],
})

plt.figure(figsize=(8,5))
plt.bar(results["Algorithm"], results["Accuracy"])
plt.xlabel("Algorithm")
plt.ylabel("Accuracy")
plt.title("Accuracy Comparison of Classification Algorithms")
plt.show()
```
